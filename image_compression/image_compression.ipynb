# Compression d'image à l'aide d'un algorithme non supervisé

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_sample_image
flower = load_sample_image('flower.jpg')
```

```python
plt.imshow(flower);
```

```python
flower.shape
```

```python
flower = flower.reshape(427*640, 3)
```

```python
flower.shape
```

```python
from sklearn.cluster import MiniBatchKMeans
km = MiniBatchKMeans(n_clusters=16)
km.fit(flower)
```

```python
km.cluster_centers_
```

```python
flower_recol = np.ceil(km.cluster_centers_[km.predict(flower)])
```

```python
flower_recol = flower_recol.reshape(427, 640, 3)
```

```python
flower_recol = flower_recol.astype('int')
```

```python
plt.imshow(flower_recol);
```
